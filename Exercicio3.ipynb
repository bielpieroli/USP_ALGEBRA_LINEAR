### Exercício 01 - Subespaço
Considere os subespaços de $\mathbb{R}^3$ dados por
$$
U = \{(x,y,z) \in \mathbb{R}^3; x = 0 \} \quad \text{e}\quad
V = \{(x,y,z) \in \mathbb{R}^3; y = 0 \}.
$$
Observe que $\mathbb{R}^3 = U + V$, porém a soma não é direta.

Implemente uma visualização adequada para $U$, $V$ e  $U \cap V$.

```python
# U é o plano x = 0
# V é o plano y = 0
# U ∩ V é a reta de interseção entre os planos, sendo ela neste caso o eixo z

# A soma não é direta, pois existe interseção!

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

lim = 100  

y = np.linspace(-lim, lim, 20)
z = np.linspace(-lim, lim, 20)
Y, Z = np.meshgrid(y, z)
X = np.zeros_like(Y)
ax.plot_surface(X, Y, Z, alpha=0.4, color='blue', label='U (x=0)')

x = np.linspace(-lim, lim, 20)
z = np.linspace(-lim, lim, 20)
X2, Z2 = np.meshgrid(x, z)
Y2 = np.zeros_like(X2)
ax.plot_surface(X2, Y2, Z2, alpha=0.4, color='green', label='V (y=0)')

ax.plot([0, 0], [0, 0], [-lim, lim], color='red', linewidth=3, label='U ∩ V (eixo z)')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Visualização dos subespaços U, V e U ∩ V em R^3')

ax.set_xlim(-lim, lim)
ax.set_ylim(-lim, lim)
ax.set_zlim(-lim, lim)


ax.legend()

plt.show()
```

#### Exercício 02 - Transformação Linear
Seja a transformação linear $T:\mathbb{R}^2 \to \mathbb{R}^2$ associada a matriz
$$ A =
\begin{bmatrix}
2 & 1 \\
1 & 2
\end{bmatrix}
.$$
Considere a circunferência $C=\{(x,y)\in\mathbb{R}^2;\; x^2+y^2=1\}$. Implemente uma visualização da imagem da circunferência $C$ pela transformação linear $T$.
- Dica: Discretize um conjunto de pontos de $C$ através da parametrização $(x, y) = (cos(t), sen(t))$ e aplique a transformação nesses pontos.

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[2, 1],
              [1, 2]])

n_points = 300
theta = np.linspace(0, 2*np.pi, n_points)
C = np.array([np.cos(theta), np.sin(theta)])

T_C = A @ C

plt.figure(figsize=(6,6))
plt.plot(C[0], C[1], label='Circunferência original (raio=1)', color='blue')
plt.plot(T_C[0], T_C[1], label='Imagem pela transformação T', color='red')

plt.gca().set_aspect('equal', adjustable='box')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Imagem da circunferência pela transformação linear T')
plt.legend()
plt.show()
```
